```python
import pandas as pd
from sklearn import datasets
```

```python
boston = datasets.load_boston()
df = pd.DataFrame(data = boston.data, columns = boston.feature_names)
```

```python
print(df.head())
```

```python
# split the dataframe
X = boston.data
y = boston.target

# X = boston.drop['MEDV', axis = 1].values
# y = boston['MEDV'].values
```

```python
# predicting the house values from a single feature: number of rooms which is column 5
X_rooms = X[:,5]
type(X_rooms), type(y)
```

```python
y = y.reshape(-1,1)
X_rooms = X_rooms.reshape(-1,1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_rooms, y)
plt.ylabel('value of house /1000 ($)')
plt.xlabel('number of rooms')
plt.show()
```

```python
# fitting a regression model
from sklearn.linear_model import LinearRegression
# create the regression model
linreg = LinearRegression()
# train the mocel
linreg.fit(X_rooms, y)
```

```python
import numpy as np
prediction_space = np.linspace(min(X_rooms), max(X_rooms)).reshape(-1,1) 
```

```python
plt.scatter(X_rooms, y, color = 'blue')
prediction = linreg.predict(prediction_space)
plt.plot(prediction_space, prediction, color = 'black', linewidth = 3)
plt.show()
```
